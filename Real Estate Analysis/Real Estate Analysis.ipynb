### Data Cleaning and Preprocessing

We have performed
1. Handle missing values
2. Remove duplicates
3. Handle outliers using z_score
4. Convert data types
5. Data transformation lower case categorical data

```python
median_price = real_estate['price'].median()
real_estate['price'].fillna(median_price,inplace=True)
median_price
```

```python
median_bed = real_estate['bed'].median()
real_estate['bed'].fillna(median_bed,inplace=True)

median_bath = real_estate['bath'].median()
real_estate['bath'].fillna(median_bath,inplace=True)

median_house = real_estate['house_size'].median()
real_estate['house_size'].fillna(median_house,inplace=True)
```

```python
real_estate.drop_duplicates(inplace=True)
```

```python
z_score = np.abs(stats.zscore(real_estate['price']))
z_score
```

```python
real_estate[(z_score>3)]
```

```python
threshold=3
real_estate=real_estate[(z_score<threshold)]
```

```python
real_estate[(real_estate['prev_sold_date']!='NaN')]
```

```python
real_estate['prev_sold_date'] = pd.to_datetime(real_estate['prev_sold_date'])
```

##### Histogram of Property Prices:

ax.set_xticklabels([f'${edge / 1000:.0f}K' for edge in bin_edges]): This line formats the x-axis tick labels to display the bin edges in thousands (K). It takes each bin edge, divides it by 1000 o convert to thousands of dollars, and formats it as a string with a '$' sign and 'K' to indicate thousands

```python
# Define the bin edges based on your price data
bin_edges = [i for i in range(0, 12500001, 2500000)]

plt.figure(figsize=(10, 6))
ax = sns.histplot(data=real_estate, x='price', bins=bin_edges)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Set x-axis tick labels to match the bin edges
ax.set_xticks(bin_edges)
ax.set_xticklabels(f'${edge/100000:.0f}Lacks'for edge in bin_edges) # ax.set_xticklabels([f'${edge / 1000:.0f}K' for edge in bin_edges]): This line formats the x-axis tick labels to display the bin edges in thousands (K). It takes each bin edge, divides it by 100000 to convert to thousands of dollars, and formats it as a string with a '$' sign and 'K' to indicate thousands
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Adjust the y-axis limits
ax.set_ylim(0, 160000)  # Set the y-axis limit based on your data
plt.xlim(0,max(bin_edges))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

##### Correlation Heatmap

```python
data=real_estate.select_dtypes(include='number')
plt.figure(figsize=(10, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

##### Bar Plot of Property Status

```python
plt.figure(figsize=(10,6))
sns.countplot(data=real_estate,x='status',palette='Set2')
plt.title('Distribution of Status')
plt.xlabel('Status')
plt.ylabel('Count')
plt.show()
```

##### Scatter plot of house size v/S Proce

```python
plt.figure(figsize=(10, 6))
plt.scatter(real_estate['house_size'], real_estate['price'], alpha=0.7, color='blue')
plt.title('Distribution based on House Size vs. Price')
plt.xlabel('Log House size')
plt.ylabel('Price')
plt.xscale('log')  # Set x-axis to logarithmic scale
plt.show()
```
